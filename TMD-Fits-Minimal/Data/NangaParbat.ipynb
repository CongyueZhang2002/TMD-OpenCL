```python
abs_y_list = [
    "CMS_7TeV", "CMS_8TeV", 
    "CMS_13TeV_y_0_0.4", "CMS_13TeV_y_0.4_0.8", "CMS_13TeV_y_0.8_1.2", "CMS_13TeV_y_1.2_1.6", "CMS_13TeV_y_1.6_2.4",
    "ATLAS_7TeV_y_0_1", "ATLAS_7TeV_y_1_2", "ATLAS_7TeV_y_2_2.4",
    "ATLAS_8TeV_Q_46_66", "ATLAS_8TeV_Q_116_150",
    "ATLAS_8TeV_y_0_0.4", "ATLAS_8TeV_y_0.4_0.8", "ATLAS_8TeV_y_0.8_1.2", "ATLAS_8TeV_y_1.2_1.6", "ATLAS_8TeV_y_1.6_2", "ATLAS_8TeV_y_2_2.4",
    "ATLAS_13TeV"
             ]
```

```python
for experiment in experiments:

    exp_dir = before_dir / "DY" / experiment
    file_names = [file.stem for file in sorted(exp_dir.glob("*.csv"))]
    file_paths = [exp_dir / f"{name}.csv" for name in file_names] 

    save_dir = Path(after_dir) / "DY" / experiment
    save_dir.mkdir(parents=True, exist_ok=True)

    for file_name in file_names:
        if file_name in includes:
            file_path = exp_dir / f"{file_name}.csv"

            df = pd.read_csv(file_path)
            
            N = len(df["xsec"])
            factor = 1
            if file_name in abs_y_list:
                factor = factor*2
            if df["observable"][0] == "1/σ*dσ/dqT":
                factor = factor/df["total_xsec"][0]

            NangaParbat_xsec_raw = NangaParbat_list[file_name][0:N]
            df["NangaParbat_xsec"] = factor*np.array(NangaParbat_xsec_raw)

            destination = after_dir / "DY" / experiment / f"{file_name}.csv"
            if destination.exists():
                raise FileExistsError(f"{file_name} already exists")
            df.to_csv(destination, index=False)
            
            print(f"{file_name} generated")
```
