# Read `emcee_gated_6d.h5`

This notebook opens the HDF5 file and prints groups/datasets with shapes.

```python
from pathlib import Path
import h5py
```

```python
h5_path = Path("replicas_np0112_scale0p01_band5 (1).h5")
if not h5_path.exists():
    raise FileNotFoundError(f"Cannot find: {h5_path.resolve()}")

print(f"Reading: {h5_path.resolve()}")
```

```python
def describe_h5(name, obj):
    if isinstance(obj, h5py.Group):
        print(f"[GROUP]   {name}")
    elif isinstance(obj, h5py.Dataset):
        print(f"[DATASET] {name} shape={obj.shape} dtype={obj.dtype}")

with h5py.File(h5_path, "r") as f:
    print("Root keys:", list(f.keys()))
    f.visititems(describe_h5)
```

```python
# Optional: preview small slices from common emcee datasets if present
with h5py.File(h5_path, "r") as f:
    if "mcmc" in f:
        mcmc = f["mcmc"]
        for key in ["chain", "log_prob", "blobs"]:
            if key in mcmc:
                ds = mcmc[key]
                print(f"\nPreview mcmc/{key}: shape={ds.shape}, dtype={ds.dtype}")
                if ds.size > 0:
                    print(ds[tuple(slice(0, min(2, d)) for d in ds.shape)])
```
